```python
#importing all the packages and dependecies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#reading the dataset
credit_card_data= pd.read_csv("C:/Users/GONE SAIKUMAR/OneDrive/Desktop/Sai project/creditcard.csv")
```

```python
#total number of normal and fraud transactions
credit_card_data['Class'].value_counts()
```

```python
#separating normal and fraud transactions
normal= credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
#after separating normal and fraud txns we are displaying the shape of data
print(normal.shape)
print(fraud.shape)
```

```python
#statistical measures of normal data
normal.Amount.describe()
```

```python
#statistical measures of fraud data
fraud.Amount.describe()
```

```python
#bar graph between fraud and normal txns
count_classes = pd.value_counts(credit_card_data['Class'], sort = True)
count_classes.plot(kind = 'bar', rot=0)
LABELS = ["Normal", "Fraud"]
plt.title("Transaction Class Distribution")
plt.xticks(range(2), LABELS)
plt.xlabel("Class")
plt.ylabel("Frequency")
```

```python
#comparing the values of normal and fraud txns
credit_card_data.groupby('Class').mean()
```

```python
#given data is unbalanced so we will select a random sample of 492 normal txns
normal_sample = normal.sample(n=492)
```

```python
#comibining two data frames(normal_sample & fraud),axis=0 is used for rows
new_dataset = pd.concat([normal_sample, fraud], axis=0)
```

```python
#displaying top 5 data from new_dataset
new_dataset.head()
```

```python
#displaying bottom 5 data from new_dataset
new_dataset.tail()
```

```python
#separating fraud and normal txns
new_dataset['Class'].value_counts()
```

```python
#comparing the values for normal_sample and fraud txns
new_dataset.groupby('Class').mean()
```

```python
#splitting data into features(columns)&targets(0&1)
X= new_dataset.drop(columns='Class',axis=1)
Y= new_dataset['Class']
```

```python
X
```

```python
Y
```

```python
#splitting data into training and testing
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn import svm
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

```python
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
training_data_accuracy
```

```python
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
test_data_accuracy
```
